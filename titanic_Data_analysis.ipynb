```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load data directly from URL
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
```

```python
# Preview first 5 rows
print(df.head())
```

```python

# Check info about dataset
print(df.info())
```

```python
# Summary statistics
print(df.describe())
```

```python
#Selects the required columns
selected_cols = ['Survived', 'Pclass' , 'Sex', 'Age']
df_selected = df[selected_cols]
print(df_selected.head())
```

```python
#filtering rows where where age is missing
df_clean = df_selected.dropna(subset=['Age'])
print(f"Number of rows after cleaning: {len(df_clean)}")
print(f"Number of rows before cleaning: {len(df_selected)}")
```

```python
# Group by Sex and Survived, count occurrences
visual = df_clean.groupby(['Sex', 'Survived']).size().unstack()
print(visual)

#plotting a bar graph
visual.plot(kind='bar', stacked=True)
plt.xlabel('Genders')
plt.ylabel('Count')
plt.title('Number of survivers by Genders')
plt.show()
```

```python
# Plot histogram of ages
plt.figure(figsize=(10, 5))
plt.hist(df_clean['Age'], bins=25, edgecolor='black')
plt.title('Age Distribution of Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.grid(True)
plt.show()
```
